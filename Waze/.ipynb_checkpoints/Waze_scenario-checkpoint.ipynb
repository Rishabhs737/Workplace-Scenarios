### **Null values and summary statistics**

Compare the summary statistics of the 700 rows that are missing labels with summary statistics of the rows that are not missing any values.

**Question:** Is there a discernible difference between the two populations?

```python
missing_labels =df[df['label'].isnull()]
missing_labels.describe()
```

```python
has_label = df[df['label'].notnull()]
has_label.describe()
```

### **Null values - device counts**

Next, check the two populations with respect to the `device` variable.

**Question:** How many iPhone users had null values and how many Android users had null values?

```python
# Get count of null values by device
missing_labels['device'].value_counts()
```

```python
# Calculate % of iPhone nulls and Android nulls
missing_iphone = missing_labels[missing_labels['device'] == 'iPhone']
missing_android = missing_labels[missing_labels['device'] == 'Android'] 

print("Iphone : " + str(round(len(missing_iphone) / len(missing_labels),2) * 100) + "%")
print("Android : " + str(round(len(missing_android) / len(missing_labels),2) * 100) + "%" )
```

```python
# Calculate % of iPhone users and Android users in full dataset
has_iphone = has_label[has_label['device'] == 'iPhone']
has_android = has_label[has_label['device'] == 'Android']
print("Iphone : " + str(round(len(has_iphone) / len(has_label),2) * 100) + "%")
print("Android : " + str(round(len(has_android) / len(has_label),2) * 100) + "%" )
```

```python
# Calculate counts of churned vs. retained
has_label['label'].value_counts(normalize=True) * 100 #TODO: should use normalize in the above query as well
```

Next, compare the medians of each variable for churned and retained users. The reason for calculating the median and not the mean is that you don't want outliers to unduly affect the portrayal of a typical user. Notice, for example, that the maximum value in the driven_km_drives column is 21,183 km. That's more than half the circumference of the earth!

```python
chur = has_label[has_label['label'] == 'churned']
retain = has_label[has_label['label'] == 'retained']

chur.median(numeric_only=True)
```

```python
retain.median(numeric_only=True)
```

```python
# Add a column to df called `km_per_drive`
df['km_per_drive'] = df['driven_km_drives'] / df['drives']
```

```python
# Group by `label`, calculate the median, and isolate for km per drive
df.groupby('label')['km_per_drive'].median(numeric_only=True)
```

```python
# Add a column to df called `km_per_driving_day`
df['km_per_driving_days'] = df['driven_km_drives']/df['driving_days']
```

```python
# Group by `label`, calculate the median, and isolate for km per driving day
df.groupby('label')['km_per_driving_days'].median(numeric_only=True)
```

```python
# For each label, calculate the number of Android users and iPhone users
# For each label, calculate the percentage of Android users and iPhone users
df.groupby('label')['device'].value_counts(normalize=True).unstack() * 100
```
